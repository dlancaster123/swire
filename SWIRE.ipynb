```python
# prompt: I have a csv file in the main documents folder called swire.csv, import this into my program using pandas dataframes

import pandas as pd
import plotly.express as px

# Assuming the file is in the main documents folder, you can use the following path:
df = pd.read_csv('swire.csv', header=2)

print(df)

def piecharts(subset_df):
    fig1 = px.pie(subset_df, values='Base', names=subset_df['Category'], title=f'Base Level Shelves in {input1} Region')
    fig2 = px.pie(subset_df, values='Signature', names=subset_df['Category'], title=f'Signature Level Shelves in {input1} Region')
    fig3 = px.pie(subset_df, values='Elite', names=subset_df['Category'], title=f'Elite Level Shelves in {input1} Region')
    fig1.show()
    fig2.show()
    fig3.show()




def nonswire(subset_df):
    print("\n")
    print(f"So what you will see is the first chart is just showing how the companies should break down their shleves, regardless of Coca-Cola products. Then the three following charts are showing what percentage of the broken down percentage (first chart) Swire is asking for.")
    fig = px.pie(subset_df, values='Shelves', names=subset_df['Category'], title=f'Shelves in {input1} Region (Not Swire)')
    fig.show()



def rows(input1, input2):
    if input1.lower()=="az and nm" and input2.lower()=="no":
        return df[0:10]
    elif input1.lower()=="oregon" and input2.lower()=="no":
        return df[10:20]
    elif input1.lower()=="oregon" and input2.lower()=="yes":
        return df[20:30]
    elif input1.lower()=="rm" and input1.lower()=="no":
         return df[30:40]
    elif input1.lower()=="rm" and input2.lower()=="yes":
        return df[40:50]
    elif input1.lower()=="northern co" and input2.lower()=="no":
        return df[50:60]
    elif input1.lower()=="northern co" and input2.lower()=="yes":
        return df[60:70]
    elif input1.lower()=="southern co" and input2.lower()=="yes":
        return df[70:80]
    elif input1.lower()=="utah" and input2.lower()=="no":
        return df[80:90]
    elif input1.lower()=="utah" and input2.lower()=="yes":
        return df[90:100]
    elif input1.lower()=="washington" and input2.lower()=="no":
        return df[100:110]
    elif input1.lower()=="eastern wa" and input2.lower()=="no":
        return df[110:120]
    elif input1.lower()=="reno" and input2.lower()=="no":
        return df[120:130]
    else:
        return df[0:0]





while True:

    input1=input("What region would you like to see? say 'quit' if you don't wish to continue")
    if input1.lower=="quit":
        break
    input2=input("Is it KO and DP?")

    subset_df = rows(input1, input2)
    if subset_df.size==0:
        print("Not Valid")
        continue
    nonswire(subset_df)
    piecharts(subset_df)
    question=input("Would you like to see another region?")
    if question.lower()=="no":
        print("I hope this has helped you visualize the information in the table!")
        break

```
